Combined life expectance and expenditure trends

```python
import matplotlib.pyplot as plt

# Create the figure and axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Health Expenditure on the left y-axis
for country in iris_df['Country'].unique():
    country_data = iris_df[iris_df['Country'] == country]
    ax1.plot(country_data['Year'], country_data['Health_Expenditure_USD'], label=f"{country} - Expenditure", linestyle="--")

ax1.set_xlabel("Year")
ax1.set_ylabel("Health Expenditure (USD)", color="tab:blue")
ax1.tick_params(axis="y", labelcolor="tab:blue")

# Create a second y-axis for Life Expectancy
ax2 = ax1.twinx()
for country in iris_df['Country'].unique():
    country_data = iris_df[iris_df['Country'] == country]
    ax2.plot(country_data['Year'], country_data['Life_Expectancy'], label=f"{country} - Life Exp.", linestyle="-")

ax2.set_ylabel("Life Expectancy (Years)", color="tab:green")
ax2.tick_params(axis="y", labelcolor="tab:green")

# Title
plt.title("Health Expenditure and Life Expectancy Trends by Country")

# Combine legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper left", fontsize="small")

plt.tight_layout()
plt.show()
```
